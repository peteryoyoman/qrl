## 選哪個動作是和環境independent的，四方向equally superposition

```python
#創建quantum circuit
dev = qml.device('default.qubit', wires = 22, shots = 10)

#'environment_0','environment_1','environment_2','environment_3','first action_0',
                  #first_action_1','second_action_0','second_action_1','third_action_0','third_action1',
                  #fourth_action_0','fourth_action_1','fifth_action_0','fifth_action_1','status_0',
                   #status_1','status_2','status_3','reward_0','reward_1','reward_2','reward3'
#對應的qubit作用0~22
            
@qml.qnode(dev)
def circuit():
    #使每一個action選擇都是equally superposition
    for i in range(10):
        qml.Hadamard(i+4)
    #使reward一開始是最大的狀態
    for j in range(4):
        qml.PauliX(j+18)
    #控制上下左右
    for k in range(2):
        qml.ControlledQubitUnitary(Go_Right, control_wires = [2*k+4,2*k+5,14,15,16,17], wires = [2,3], control_values='000000')
        qml.ControlledQubitUnitary(Go_Left, control_wires = [2*k+4,2*k+5,14,15,16,17], wires = [2,3], control_values='100000')
        qml.ControlledQubitUnitary(Go_Up, control_wires = [2*k+4,2*k+5,14,15,16,17], wires = [0,1], control_values='010000')
        qml.ControlledQubitUnitary(Go_Down, control_wires = [2*k+4,2*k+5,14,15,16,17], wires = [0,1], control_values='110000')
        #判斷踩到什麼東西去判斷怎麼操作end_status
        #Part 1:踩到洞
        qml.ControlledQubitUnitary(end_status_1, control_wires = [0,1,2,3], wires = [14,15,16,17], control_values='0101')
        qml.ControlledQubitUnitary(end_status_1, control_wires = [0,1,2,3], wires = [14,15,16,17], control_values='0111')
        qml.ControlledQubitUnitary(end_status_1, control_wires = [0,1,2,3], wires = [14,15,16,17], control_values='1011')
        qml.ControlledQubitUnitary(end_status_1, control_wires = [0,1,2,3], wires = [14,15,16,17], control_values='1100')
        #Part 2:踩到終點
        qml.ControlledQubitUnitary(end_status_2, control_wires = [0,1,2,3], wires = [14,15,16,17], control_values='1010')

        #由end_status qubits去判斷reward要怎麼動作
        #Part 1:什麼東西都沒踩到過
        qml.ControlledQubitUnitary(reward_unitary, control_wires = [14,15,16,17], wires = [18,19,20,21], control_values='0000')
        #Part 2:第一次踩到洞
        qml.ControlledQubitUnitary(punish_unitary, control_wires = [14,15,16,17], wires = [18,19,20,21], control_values='0001')
        #Part 3:第一次踩到終點
        qml.ControlledQubitUnitary(goal_unitary, control_wires = [14,15,16,17], wires = [18,19,20,21], control_values='1111')
    
    return [qml.sample(qml.PauliZ(i)) for i in range(22)]
```

```python
#創建quantum circuit
num_of_repeat = 3
dev = qml.device('default.qubit', wires = 12+num_of_repeat*4, shots = 20)
shape = BasicEntanglerLayers.shape(n_layers=1, n_wires=4)
initial_weights = np.random.random(size=shape)*np.pi
@qml.qnode(dev)
def circuit(weights):
    #使reward一開始是最大的狀態
    for j in range(4):
        qml.PauliX(j+((num_of_repeat*4)+8))
    #控制上下左右
    for k in range(num_of_repeat):
        qml.CNOT(wires=[0,4*k+4])
        qml.CNOT(wires=[1,4*k+5])
        qml.CNOT(wires=[2,4*k+6])
        qml.CNOT(wires=[3,4*k+7])
        BasicEntanglerLayers(weights=weights, wires=[4*k+4,4*k+5,4*k+6,4*k+7])
        qml.ControlledQubitUnitary(Go_Right, control_wires = [4*k+4,4*k+5,4*num_of_repeat+4,4*num_of_repeat+5,4*num_of_repeat+6,4*num_of_repeat+7], wires = [2,3], control_values='000000')
        qml.ControlledQubitUnitary(Go_Left, control_wires = [4*k+4,4*k+5,4*num_of_repeat+4,4*num_of_repeat+5,4*num_of_repeat+6,4*num_of_repeat+7], wires = [2,3], control_values='100000')
        qml.ControlledQubitUnitary(Go_Up, control_wires = [4*k+4,4*k+5,4*num_of_repeat+4,4*num_of_repeat+5,4*num_of_repeat+6,4*num_of_repeat+7], wires = [0,1], control_values='010000')
        qml.ControlledQubitUnitary(Go_Down, control_wires = [4*k+4,4*k+5,4*num_of_repeat+4,4*num_of_repeat+5,4*num_of_repeat+6,4*num_of_repeat+7], wires = [0,1], control_values='110000')

        #判斷踩到什麼東西去判斷怎麼操作end_status
        #Part 1:踩到洞
        qml.ControlledQubitUnitary(end_status_1, control_wires = [0,1,2,3], wires = [4*num_of_repeat+4,4*num_of_repeat+5,4*num_of_repeat+6,4*num_of_repeat+7], control_values='0101')
        qml.ControlledQubitUnitary(end_status_1, control_wires = [0,1,2,3], wires = [4*num_of_repeat+4,4*num_of_repeat+5,4*num_of_repeat+6,4*num_of_repeat+7], control_values='0111')
        qml.ControlledQubitUnitary(end_status_1, control_wires = [0,1,2,3], wires = [4*num_of_repeat+4,4*num_of_repeat+5,4*num_of_repeat+6,4*num_of_repeat+7], control_values='1011')
        qml.ControlledQubitUnitary(end_status_1, control_wires = [0,1,2,3], wires = [4*num_of_repeat+4,4*num_of_repeat+5,4*num_of_repeat+6,4*num_of_repeat+7], control_values='1100')
        #Part 2:踩到終點
        qml.ControlledQubitUnitary(end_status_2, control_wires = [0,1,2,3], wires = [4*num_of_repeat+4,4*num_of_repeat+5,4*num_of_repeat+6,4*num_of_repeat+7], control_values='0010')

        #由end_status qubits去判斷reward要怎麼動作
        #Part 1:什麼東西都沒踩到過
        qml.ControlledQubitUnitary(reward_unitary, control_wires = [4*num_of_repeat+4,4*num_of_repeat+5,4*num_of_repeat+6,4*num_of_repeat+7], wires = [4*num_of_repeat+8,4*num_of_repeat+9,4*num_of_repeat+10,4*num_of_repeat+11], control_values='0000')
        #Part 2:第一次踩到洞
        qml.ControlledQubitUnitary(punish_unitary, control_wires = [4*num_of_repeat+4,4*num_of_repeat+5,4*num_of_repeat+6,4*num_of_repeat+7], wires = [4*num_of_repeat+8,4*num_of_repeat+9,4*num_of_repeat+10,4*num_of_repeat+11], control_values='0001')
        #Part 3:第一次踩到終點
        qml.ControlledQubitUnitary(goal_unitary, control_wires = [4*num_of_repeat+4,4*num_of_repeat+5,4*num_of_repeat+6,4*num_of_repeat+7], wires = [4*num_of_repeat+8,4*num_of_repeat+9,4*num_of_repeat+10,4*num_of_repeat+11], control_values='1111')
    return [qml.sample(qml.PauliZ(i)) for i in range(num_of_repeat*4+12)]
```

```python
def cost(params):
    y = circuit(params)
    #y = np.where(x==1,0,1)
    y1 = y[20:24]
    print(y1)
    total_reward = y1[0]*8 + y1[1]*4 + y1[2]*2 + y1[3]
    return total_reward

cost(initial_weights)
```

```python
opt = qml.AdamOptimizer(stepsize=0.2)
steps = 100
weights = initial_weights
for i in range(steps):
    weights = opt.step(cost, weights)
    #print("Now angles: {}".format(weights))
    
    if (i + 1) % 5 == 0:
        print("Cost after step {:5d}: {: .7f}".format(i + 1, cost(weights)))

print("Optimized rotation angles: {}".format(weights))
```
